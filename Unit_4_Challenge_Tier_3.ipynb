```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transposed the dataframe
london_df = london_df.transpose()
london_df.head()
```

```python
# Assigned the first row as columns
london_df.columns = london_df.iloc[0]

london_df.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Renamed the column names
london_df = london_df.rename(columns={'Unnamed: 0' : 'london_bor', pd.NaT :'ID'})
london_df.head()

```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melted the dataframe with month and avg_price columns
london_df = pd.melt(london_df, id_vars=['london_bor', 'ID'], var_name='month', value_name='avg_price')
london_df.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Dropped the first row as it contained null values
london_df = london_df.drop(0)
london_df.head()
```

```python
# Checked the datatypes of columns
london_df.dtypes
```

```python
# Replaced all missing values to NaN format in the avg_price column
london_df['avg_price'].replace('', np.nan, inplace=True)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Dropped all null values in avg_price & ID columns
london_df.dropna(subset=['avg_price'], inplace=True)
london_df.dropna(subset=['ID'], inplace=True)
```

```python
# Checked for any null values
london_df.isna().any()
```

```python
# Checked dataframe for equal counts in columns
london_df.count()
```

```python
# Converted avg_price datatype into float
london_df['avg_price'] = london_df['avg_price'].astype(float)
```

```python
# Checked for the updated datatypes
london_df.dtypes
```

```python
# Checked the uniques borough names in the column
london_df['london_bor'].unique()
```

```python
# Removing the non borough names from dataframe
# A list of non-boroughs. 
nonBoroughs = ['City of London', 'Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']

# Checked the non boroughs list in the dataframe
london_df[london_df.london_bor.isin(nonBoroughs)]

# Excluded the non boroughs list & retrieved only the true boroughs
london_df[~london_df.london_bor.isin(nonBoroughs)]

# Assigned the result as clean_london 
clean_london = london_df[~london_df.london_bor.isin(nonBoroughs)]
clean_london.head(35)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Visualized data with line plot for all the london boroughs
clean_london.plot(x="month", y="avg_price", kind="line", rot=90)

# Visualized data only for 'Brent' borough
city_london = clean_london[clean_london["london_bor"] == "Brent"]

city_london.plot(x="month", y="avg_price", kind="line", rot=90)
```

```python
# Grouped the dataframe by year and borough names, with mean function
clean_london = clean_london.groupby(["london_bor", "year"])["avg_price"].mean()
clean_london = clean_london.reset_index()

clean_london.head(10)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Function to calculate ratio
def create_price_ratio(a):
  y1998 = float(a['avg_price'][a['year'] == 1998])
  y2018 = float(a['avg_price'][a['year'] == 2018])
  ratio = [y1998/y2018]
  return ratio
```

```python
# Testing the function with one borough name
create_price_ratio(clean_london[clean_london['london_bor']=='Camden'])
```

```python
# Creating a empty dictionary to hold final result
final = {}
```

```python
# Writing a for loop to iterate over entire dataframe
for b in clean_london['london_bor'].unique():
  borough = clean_london[clean_london['london_bor'] == b]
  final[b] = create_price_ratio(borough)
print(final)
```

```python
# Converting the final dictionary into dataframe
final_df = pd.DataFrame(final)

# Applying transpose
final_df = final_df.transpose()

# Resetting index
final_df = final_df.reset_index()

# Updating column names
final_df = final_df.rename(columns = {'index' : 'London boroughs', 0 : 'Ratios'})

# Sorting the ratios from highest to lowest
final_df = final_df.sort_values(by='Ratios', ascending=False)
final_df.head(10)
```

```python
# Visualizing the ratio of london boroughs

final_plot = final_df.plot(x='London boroughs', y='Ratios', kind='bar', rot=90)
print(final_plot)
```
